#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
def map_age_range(age_str):
    if pd.isnull(age_str):
        return None
    elif 'Under 18' in age_str:
        return 15
    elif '18-24' in age_str:
        return 21
    elif '25-34' in age_str:
        return 29.5
    elif '35-44' in age_str:
        return 39.5
    elif '45-54' in age_str:
        return 49.5
    elif '55-64' in age_str:
        return 59.5
    elif '65 or older' in age_str or '65+' in age_str:
        return 67.5
    else:
        return None

df['AgeNumeric'] = df['Age'].apply(map_age_range)

new_data = df[['AgeNumeric','JobSatPoints_6']].dropna()

plt.figure(figsize=(7, 5))
sns.scatterplot(data=new_data, x='JobSatPoints_6', y='AgeNumeric', alpha=0.5)
plt.title('Relationship Between Age of Respondents and Their Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Respondents Age')
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
new_data = df[['ConvertedCompYearly','JobSatPoints_6']].dropna()

plt.figure(figsize=(7, 5))
sns.scatterplot(data=new_data, x='JobSatPoints_6', y='ConvertedCompYearly', alpha=0.5)
plt.title('ConvertedCompYearly vs Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Yearly Compensation')
plt.yscale('log')
plt.tight_layout()
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
!pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
```

```python
## Write your code here
def map_age_range(age_str):
    if pd.isnull(age_str):
        return None
    elif 'Under 18' in age_str:
        return 15
    elif '18-24' in age_str:
        return 21
    elif '25-34' in age_str:
        return 29.5
    elif '35-44' in age_str:
        return 39.5
    elif '45-54' in age_str:
        return 49.5
    elif '55-64' in age_str:
        return 59.5
    elif '65 or older' in age_str or '65+' in age_str:
        return 67.5
    else:
        return None

df['AgeNumeric'] = df['Age'].apply(map_age_range)

new_data = df[['AgeNumeric','JobSatPoints_6']].dropna()

lm=LinearRegression()
x=new_data[['JobSatPoints_6']]
y=new_data[['AgeNumeric']]
lm.fit(x,y)
plt.figure(figsize=(7, 5))
sns.scatterplot(data=new_data, x='JobSatPoints_6', y='AgeNumeric', alpha=0.5)
plt.title('Relationship Between Age of Respondents and Their Job Satisfaction')
y_pred = lm.predict(x)
plt.plot(new_data['JobSatPoints_6'], y_pred, color='red', label='Regression Line')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Respondents Age')
plt.tight_layout()
plt.legend()
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
## Write your code here
def map_age_range(age_str):
    if pd.isnull(age_str):
        return None
    elif 'Under 18' in age_str:
        return 15
    elif '18-24' in age_str:
        return 21
    elif '25-34' in age_str:
        return 29.5
    elif '35-44' in age_str:
        return 39.5
    elif '45-54' in age_str:
        return 49.5
    elif '55-64' in age_str:
        return 59.5
    elif '65 or older' in age_str or '65+' in age_str:
        return 67.5
    else:
        return None

df['AgeNumeric'] = df['Age'].apply(map_age_range)

df['YearsCodePro'] = df['YearsCodePro'].replace({'Less than 1 year': 0.5,'More than 50 years': 51})
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

new_df=df[['AgeNumeric','YearsCodePro']].dropna()

plt.figure(figsize=(7, 5))
sns.scatterplot(data=new_df, x='AgeNumeric', y='YearsCodePro', alpha=0.5)
plt.title('Relationship Between Age of Respondents and Their Work Experience')
plt.xlabel('Respondents Age')
plt.ylabel('Work Experience')
plt.tight_layout()
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here
def map_age_range(age_str):
    if pd.isnull(age_str):
        return None
    elif 'Under 18' in age_str:
        return 15
    elif '18-24' in age_str:
        return 21
    elif '25-34' in age_str:
        return 29
    elif '35-44' in age_str:
        return 39
    elif '45-54' in age_str:
        return 49
    elif '55-64' in age_str:
        return 59
    elif '65 or older' in age_str or '65+' in age_str:
        return 67
    else:
        return None

df['AgeNumeric'] = df['Age'].apply(map_age_range)

df_clean = df[['ConvertedCompYearly', 'JobSatPoints_6', 'AgeNumeric']].dropna()
df_clean = df_clean[df_clean['ConvertedCompYearly'] < 500000]

plt.figure(figsize=(10, 7))
plt.scatter(
    df_clean['ConvertedCompYearly'],
    df_clean['JobSatPoints_6'],
    s=df_clean['AgeNumeric'],
    alpha=0.5,
    c=df_clean['AgeNumeric'],
    cmap='viridis',
    edgecolors='w'
)
plt.xlabel('Yearly Compensation')
plt.ylabel('Job Satisfaction (Scale 1100)')
plt.title('Bubble Plot: Compensation vs. Job Satisfaction (Bubble Size/Color = Age)')
plt.colorbar(label='Age')
plt.grid(True)
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
df_lang = df[['JobSat', 'LanguageHaveWorkedWith']].dropna()
df_lang['LanguageHaveWorkedWith'] = df_lang['LanguageHaveWorkedWith'].str.split(';')
df_exploded = df_lang.explode('LanguageHaveWorkedWith')

top_langs = df_exploded['LanguageHaveWorkedWith'].value_counts().head(10).index
df_top = df_exploded[df_exploded['LanguageHaveWorkedWith'].isin(top_langs)]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_top, x='LanguageHaveWorkedWith', y='JobSat', alpha=0.5)
plt.title('Scatter Plot: Programming Languages vs. Job Satisfaction')
plt.xlabel('Programming Language')
plt.ylabel('Job Satisfaction (Scale 110)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
df_emp = df[['ConvertedCompYearly', 'JobSatPoints_6', 'Employment']].dropna()
df_emp = df_emp[df_emp['ConvertedCompYearly'] < 500000]

plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df_emp,
    x='ConvertedCompYearly',
    y='JobSatPoints_6',
    hue='Employment',
    alpha=0.6
)
plt.title('Scatter Plot: Compensation vs. Job Satisfaction by Employment Type')
plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction (Scale 110)')
plt.legend(title='Employment Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here

def map_age_range(age_str):
    if pd.isnull(age_str):
        return None
    elif 'Under 18' in age_str:
        return 15
    elif '18-24' in age_str:
        return 21
    elif '25-34' in age_str:
        return 29
    elif '35-44' in age_str:
        return 39
    elif '45-54' in age_str:
        return 49
    elif '55-64' in age_str:
        return 59
    elif '65 or older' in age_str or '65+' in age_str:
        return 67
    else:
        return None

df['AgeNumeric'] = df['Age'].apply(map_age_range)

df_exp = df[['AgeNumeric', 'YearsCodePro', 'Country']].dropna()

df['YearsCodePro'] = df['YearsCodePro'].replace({'Less than 1 year': 0.5,'More than 50 years': 51})
df_exp['YearsCodePro'] = pd.to_numeric(df_exp['YearsCodePro'], errors='coerce')
df_exp = df_exp.dropna()

top_countries = df_exp['Country'].value_counts().head(5).index
df_filtered = df_exp[df_exp['Country'].isin(top_countries)]


plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df_filtered,
    x='AgeNumeric',
    y='YearsCodePro',
    hue='Country',
    style='AgeNumeric',
    alpha=0.6
)

plt.title('Work Experience (YearsCodePro) Across Age Groups and Countries')
plt.xlabel('Age Group')
plt.ylabel('Years of Professional Coding Experience')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Country')
plt.grid(True)
plt.tight_layout()
plt.show()
```
